```python
# Scraping the Wikipedia page
from bs4 import BeautifulSoup
import requests
html_string = requests.get('http://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M').text
dfs = pd.read_html(html_string, header=0)
df = dfs[0]
df.head()
```

```python
# Ignoring cells with a borough that is 'Not assigned'
prog_a = df[df.Borough != 'Not assigned']
prog_a
```

```python
# Combining 'Neighbourhood' into one row with the neighborhoods for one postal code every, separated with a comma
prog_b = prog_a.groupby(['Postcode', 'Borough'])['Neighbourhood'].apply(', '.join).reset_index()
prog_b
```

```python
# Changing the value "Not assigned" in colummn 'Neighbourhood' for value of colummn 'Borough'
mask = (prog_b['Neighbourhood']=='Not assigned')
prog_b.loc[mask, 'Neighbourhood'] = prog_b['Borough']
prog_b
```

```python
import numpy as np  
import pandas as pd 
```

```python
# lets get the csv file because of changes in Google Maps API policy
df_coord = pd.read_csv('https://cocl.us/Geospatial_data')
df_coord.head ()
```

```python
# now merge two dataframes and cut the row 'Postal Code' as it copy of 'Postcode'
result = pd.merge(left=prog_b, right=df_coord, how='left', left_on='Postcode', right_on='Postal Code')
result = result.drop ('Postal Code', axis = 1)
result.head ()
```

```python
# now, lets get the Toronto latitude and longitude
address = 'Toronto, ON'
geolocator = Nominatim(user_agent="ny_explorer")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
!conda install -c conda-forge folium=0.5.0 --yes 
import folium # map rendering library
```

```python
# create map of Toronto
map_tor = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
for lat, lng, borough, neighborhood in zip(result['Latitude'],result['Longitude'], result['Borough'], result['Neighbourhood']):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color='blue',
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_tor)  
map_tor
```
